```python
x = torch.randn(3)
x = Variable(x,requires_grad = True)

y = x * 2

while y.data.norm() < 1000:
    y = y *2
print(x.data)

print(y)
```

```python
# y.backward(torch.FloatTensor(0.1))
# print(x.grad)

gradients = torch.FloatTensor([0.1,1.0,0.0001])
y.backward(gradient=gradients)
print(x.grad)
```
